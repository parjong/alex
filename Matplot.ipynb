```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Draw Historgram
#
# References
# - https://wikidocs.net/92112
# - https://numpy.org/doc/stable/reference/random/generated/numpy.random.chisquare.html
# - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html
DATA = np.random.chisquare(1.0, size=100)

fig, ax = plt.subplots()
# Q) Which xrange is proper for impact assessement?
#
# Design No 1. Baseline Interval to Error Ratio
# 
# What happens if Error > Interval..?
#
# Design No 2. Percentiles
#
# 0-th percentile ~ 1-th percentile ... 99-th percentile ~ 100-th percentile
#
# Pros. This metric is relatively stable.
# Cons. This metric is improper to visualize how severe the error is
ax.hist(DATA, bins=100)
# Which bins, 10? 100
```

```python
# Draw Heatmap
#
# References
# - https://rfriend.tistory.com/419
# - https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html
H = 16
W = 48

HM_DATA = np.random.chisquare(1.0, size=(H, W))

fig, ax = plt.subplots()

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.matshow.html
# https://stackoverflow.com/questions/38973868/adjusting-gridlines-and-ticks-in-matplotlib-imshow
im = ax.matshow(HM_DATA)

# https://stackoverflow.com/questions/12608788/changing-the-tick-frequency-on-x-or-y-axis-in-matplotlib
ax.xaxis.set_major_locator(ticker.MultipleLocator(W - 1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(H - 1))

# https://stackoverflow.com/questions/23876588/matplotlib-colorbar-in-each-subplot
fig.colorbar(im)
# Q) How to decide the color of each pixel?
```
